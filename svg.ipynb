```python
xmltodict
import json
import os
file = open("Glaucia_min.svg", "r")
xml_string = file.read()

# Parse the XML string to a Python dictionary
python_dict = xmltodict.parse(xml_string)

file.close()
print("layer2") # for the part of layer 2
print("rect")
layer2_list = python_dict['svg']['g'][1]['rect'] # 1, becuase it is second appearance of element g (count from 0)

rectangles2_data=[]

for x in range(len(layer2_list)):
    #print(layer3_list[x])
    #print(type(layer3_list[x]))
    layer2_dict = layer2_list[x]
    rectangle2_data={}

    for key, value in layer2_dict.items() :
      if key =="@style" or key == "@id" or key == "@x" or key =="@y" or key =="@width" or key =="@height":
        print(key, value)
        rectangle2_data[key] = value
    rectangles2_data.append(rectangle2_data)

print("Ellipses")
layer3_list = python_dict['svg']['g'][2]['ellipse']  # 2 because it is the third appearance of element g (count from 0)

ellipses_data=[]

for x in range(len(layer3_list)):
    layer3_dict = layer3_list[x]
    ellipse_data ={}

    for key, value in layer3_dict.items():
        if key =="@style" or key == "@id" or key == "@cx" or key == "@cy" or key == "@rx" or key == "@ry":
            print(key, value)
            ellipse_data[key] = value
    ellipses_data.append(ellipse_data)

print("rect")
layer3_list = python_dict['svg']['g'][2]['rect'] # 2, becuase it is third appearance of element g (count from 0)

rectangles_data=[]

for x in range(len(layer3_list)):
    #print(layer3_list[x])
    #print(type(layer3_list[x]))
    layer3_dict = layer3_list[x]
    rectangle_data={}

    for key, value in layer3_dict.items() :
      if key =="@style" or key == "@id" or key == "@x" or key =="@y" or key =="@width" or key =="@height":
        print(key, value)
        rectangle_data[key] = value
    rectangles_data.append(rectangle_data)

print("circle")
layer3_list = python_dict['svg']['g'][2]['circle'] # 2, becuase it is third appearance of element g (count from 0)
circles_data=[]

for x in range(len(layer3_list)):
    #print(layer3_list[x])
    #print(type(layer3_list[x]))
    layer3_dict = layer3_list[x]
    circle_data={}

    for key, value in layer3_dict.items() :
      if key =="@style" or key == "@id" or key == "@cx" or key == "@cy" or key == "@r":
        print(key, value)
        circle_data[key]= value
    circles_data.append(circle_data)

# Combine ellipses and rectangles data
combined_data = { "rectangles2": rectangles2_data, "ellipses": ellipses_data, "rectangles": rectangles_data, "circles":circles_data}

# Serializing json
json_object = json.dumps(combined_data, indent=4)

# Writing to combined_sample.json
with open("combined_sample.json", "w") as outfile:
    outfile.write(json_object)

```
